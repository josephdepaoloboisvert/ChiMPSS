```python
def print_some_info(fmazer, iter_len=1.0, timestep=2.0):
    
    if not hasattr(fmazer, 't0'):
         _ = fmazer.importance_resampling()
    
    total_iters = fmazer.energies.shape[0]
    n_states = fmazer.energies.shape[1]

    equil_iter = fmazer.t0
    percent_discarded = equil_iter/total_iters

    agg_sim_time = total_iters*n_states*iter_len #ps
    agg_sim_time /= 1e6 #make microseconds

    print(f"Simulated {n_states} states for {agg_sim_time:0.3f} aggregate microseconds.")
    print(f"\t Discarded {percent_discarded*100:2.2f}% to equilibration...")
```

```python
fast_analyzers = [fast1, fast2, fast3, fast_300_360, fast_300_400, fast_310_400]
n_resampled = 2500
```

```python
for i, (analyzer, (iterlen, timestep)) in enumerate(zip(fast_analyzers, [(7.0, 3.5)]*3 + [(14.0, 3.5)]*3)):
    print(i, analyzer, iterlen, timestep)
    _ = analyzer.importance_resampling(n_samples=n_resampled)
    print_some_info(analyzer, iter_len=iterlen, timestep=timestep)
    analyzer.plot_weights()
    analyzer.plot_energy_distributions()
```

```python
import os, sys, glob
import mdtraj as md

#sys.path.append('/expanse/lustre/projects/uil133/josephdb/Bronzeville/')
```

```python
dcds = sorted(glob.glob('/expanse/lustre/projects/uil133/josephdb/Analysis_Data/AM630_FM_Comparison/*_2.5k/resampled.dcd'))
```

```python
dcds
```

```python
pdbs = [dcd.replace('.dcd', '.pdb') for dcd in dcds]
assert False not in [os.path.isfile(pdb) for pdb in pdbs]
```

```python
proteins = []
for dcd, pdb in zip(dcds, pdbs):
    traj = md.load(dcd, top=pdb)
    proteins.append(traj.atom_slice(traj.top.select('protein')))
supertraj = proteins[0].join(proteins[1:])

supertraj = supertraj.superpose(supertraj)

supertraj[0].save_pdb('AM630_fast_comparison_2.5k_super.pdb')
supertraj.save_dcd('AM630_fast_comparison_2.5k_super.dcd')

supertraj = supertraj[::10]
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
CA_inds = supertraj.top.select('name CA')
pos = supertraj.xyz[:, CA_inds, :]
scaled_pos = (pos.max(axis=(0, 1)) - pos)/(pos.max(axis=(0, 1)) - pos.min(axis=(0,1)))
scaled_pos = scaled_pos.reshape(scaled_pos.shape[0], -1)
```

```python
comp = pca.fit_transform(scaled_pos)
```
