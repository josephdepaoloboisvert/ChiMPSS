```python
import mdtraj as md
def traj2CAcoords(dcd_fn, pdb_fn):
    traj = md.load(dcd_fn, pdb_fn)
    ca_inds = traj.top.select('protein and name CA')
    ca_traj = traj.atom_slice(ca_inds)
    return ca_traj.xyz

```

```python
resampled_from_default = md.load(['/expanse/lustre/projects/iit122/josephdb/KOR_Luis/resampled_trajectories/RS1125_0/default_pos.dcd',
                                 '/expanse/lustre/projects/iit122/josephdb/KOR_Luis/resampled_trajectories/RS1125_1/default_pos.dcd',
                                 '/expanse/lustre/projects/iit122/josephdb/KOR_Luis/resampled_trajectories/RS1125_2/default_pos.dcd'],
                                 top='/expanse/lustre/projects/iit122/josephdb/KOR_Luis/resampled_trajectories/RS1125_0/default_top.pdb')
```

```python
resampled_from_mpi = md.load(['/expanse/lustre/projects/iit122/josephdb/KOR_Luis/resampled_trajectories/RS1125_0/mpi10_pos.dcd',
                              '/expanse/lustre/projects/iit122/josephdb/KOR_Luis/resampled_trajectories/RS1125_1/mpi10_pos.dcd',
                              '/expanse/lustre/projects/iit122/josephdb/KOR_Luis/resampled_trajectories/RS1125_2/mpi10_pos.dcd'],
                              top='/expanse/lustre/projects/iit122/josephdb/KOR_Luis/resampled_trajectories/RS1125_0/mpi10_top.pdb')
```

```python
resampled_from_mpi.xyz.shape, resampled_from_default.xyz.shape
```

```python
mpi_ca_inds = resampled_from_mpi.top.select('protein and name CA')
default_ca_inds = resampled_from_default.top.select('protein and name CA')
```

```python
import numpy as np
np.all(mpi_ca_inds == default_ca_inds)
```

```python
mpi_resampled_CAs = resampled_from_mpi.atom_slice(mpi_ca_inds)
default_resampled_CAs = resampled_from_default.atom_slice(default_ca_inds)
```

```python
mpi_resampled_CAs = mpi_resampled_CAs.superpose(mpi_resampled_CAs)
default_resampled_CAs = default_resampled_CAs.superpose(default_resampled_CAs)
```

```python
mpi_pos = mpi_resampled_CAs.xyz
def_pos = default_resampled_CAs.xyz
```

```python
mpi_pos.shape, def_pos.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
def_comps = pca.fit_transform(def_pos.reshape(def_pos.shape[0], -1))
mpi_comps = pca.transform(mpi_pos.reshape(mpi_pos.shape[0], -1))
```

```python
import matplotlib.pyplot as plt

plt.scatter(def_comps[:, 0], def_comps[:, 1], label='Default', alpha=0.25)
plt.scatter(mpi_comps[:, 0], mpi_comps[:, 1], label='MPI 10', alpha=0.25)
plt.legend()
```

```python
from scipy.stats import gaussian_kde

# 1. Perform Gaussian KDE
X, Y = def_comps[:, 0], def_comps[:, 1]
xy = np.vstack([X, Y])
kernel = gaussian_kde(xy)
Z = kernel(np.vstack([X.ravel(), Y.ravel()]))

# 2. Plot
plt.contourf(X, Y, Z.reshape(X.shape), cmap='Blues')
plt.show()
```
